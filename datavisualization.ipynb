```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


wine_data = pd.read_csv('WineQT.csv')


plt.figure(figsize=(10, 6))
sns.histplot(wine_data['alcohol'], bins=20, kde=True, color='blue')
plt.title('Distribution of Alcohol Content')
plt.xlabel('Alcohol Content')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd

wine_data = pd.read_csv('WineQT.csv')

null_values = wine_data.isnull().sum()
print("Null values in each column:\n", null_values)

wine_data_cleaned = wine_data.dropna()

if 'alcohol' in wine_data.columns:
wine_data['alcohol'] = wine_data['alcohol'].fillna(wine_data['alcohol'].mean())

wine_data['pH'] = wine_data['pH'].fillna(0)

print("Dataset after handling null values:\n", wine_data.head())
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='quality', y='pH', data=wine_data, palette='muted')
plt.title('Distribution of pH by Quality')
plt.xlabel('Quality')
plt.ylabel('pH')
plt.show()


```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='alcohol', y='quality', data=wine_data, color='blue')
plt.title('Alcohol Content vs. Quality')
plt.xlabel('Alcohol Content')
plt.ylabel('Quality')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


data = {'type': ['Red', 'White', 'Red', 'White'],
    'acidity': [3.1, 3.5, 3.3, 3.4],
    'sugar': [0.5, 0.4, 0.6, 0.3],
    'alcohol': [12.5, 13.0, 12.0, 13.5]}

avg_features_by_type = pd.DataFrame(data).groupby('type').mean().reset_index()
features_of_interest = ['acidity', 'sugar', 'alcohol']  

for feature in features_of_interest:
    plt.figure(figsize=(8, 5))
    sns.barplot(x='type', y=feature, data=avg_features_by_type, palette='muted')
    plt.title(f'Average {feature.capitalize()} by Wine Type')
    plt.xlabel('Wine Type')
    plt.ylabel(f'Average {feature.capitalize()}')
    plt.show()

```
